```python
anomalies = [
    {
        "name": "Age Ratio Drift",
        "category": "Demographic Ratio",
        "description": "Sudden change in the ratio between age groups (demo_age_17_ / demo_age_5_17)",
        "intuition": "Demographic age ratios should be stable over time.",
        "detection_logic": "Ratio deviates > k std from rolling mean.",
        "evidence": {
            "level": "pincode",
            "columns_used": ["date", "pincode", "demo_age_17_", "demo_age_5_17"],
            "anomaly_count": 0,
            "results_df": None
        }
    },
    {
        "name": "Impossible Population Growth",
        "category": "Temporal Growth",
        "description": "Population increases faster than biologically possible.",
        "intuition": "Population should not grow sharply day-to-day.",
        "detection_logic": "Daily growth rate > threshold (e.g., 5%).",
        "evidence": {
            "level": "district",
            "columns_used": ["date", "district", "demo_age_5_17", "demo_age_17_"],
            "anomaly_count": 0,
            "results_df": None
        }
    },
    {
        "name": "Spatial-Temporal Stagnation",
        "category": "Data Quality",
        "description": "Identical population counts over multiple days (Frozen Data).",
        "intuition": "Real-world data usually has minor daily fluctuations; flatlines suggest reporting failures.",
        "detection_logic": "Standard deviation of count over N days is exactly 0.",
        "evidence": {
            "level": "pincode",
            "columns_used": ["date", "pincode", "demo_age_5_17"],
            "anomaly_count": 0,
            "results_df": None
        }
    },
    {
        "name": "Geographical Multi-Mapping",
        "category": "Geographical Integrity",
        "description": "A single pincode maps to multiple different districts or states.",
        "intuition": "Postal codes are typically nested within specific administrative districts.",
        "detection_logic": "Group by pincode and flag if unique(district) > 1.",
        "evidence": {
            "level": "pincode",
            "columns_used": ["pincode", "district", "state"],
            "anomaly_count": 0,
            "results_df": None
        }
    },
    {
        "name": "Cross-Region Population Leakage",
        "category": "Spatial Consistency",
        "description": "Simultaneous population drop in one district and spike in another district within the same state on the same day.",
        "intuition": "Population does not physically relocate across districts overnight; opposing spikes suggest reporting or aggregation issues.",
        "detection_logic": "Within a state and date, flag districts where population decreases >10% while another increases >10%.",
        "evidence": {
            "level": "district",
            "columns_used": ["date", "state", "district", "demo_age_5_17", "demo_age_17_"],
            "anomaly_count": 0,
            "results_df": None
        }
    }
]

print(f"Total anomaly types defined: {len(anomalies)}")
```

```python
import pandas as pd
import numpy as np

# 1. Clean slate: Remove columns from previous runs
cols_to_drop = [
    'pincode_mean', 'pincode_std', 'age_ratio', 'ratio_zscore',
    'growth_rate', 'prev_count', 'rolling_std',
    'district_total', 'district_growth'
]

df_demographic = df_demographic.drop(
    columns=[c for c in cols_to_drop if c in df_demographic.columns],
    errors='ignore'
)

epsilon = 1e-6

# ------------------------------------------------
# ANOMALY 1 (REVISED): Age Ratio Drift using District-Level Baseline
# ------------------------------------------------

# Step 1: Compute age ratio
df_demographic['age_ratio'] = (
    df_demographic['demo_age_17_'] /
    (df_demographic['demo_age_5_17'] + epsilon)
)

# Step 2: Compute district-level distribution parameters
district_stats = (
    df_demographic
    .groupby('district')['age_ratio']
    .agg(['mean', 'std', 'count'])
    .reset_index()
)

district_stats.columns = [
    'district',
    'district_mean',
    'district_std',
    'district_count'
]

# Step 3: Merge district statistics back
df_temp = df_demographic.merge(
    district_stats,
    on='district',
    how='left'
)

# Step 4: Compute district-based z-score
df_temp['ratio_zscore'] = (
    (df_temp['age_ratio'] - df_temp['district_mean']) /
    (df_temp['district_std'] + epsilon)
)

# Step 5: Apply statistically defensible threshold
# 3.0 corresponds to ~99.7% coverage under normal assumptions
drift_flags = df_temp[
    (df_temp['district_count'] >= 30) &      # ensure sufficient sample size
    (df_temp['ratio_zscore'].abs() > 3.0)
].copy()

# Store results
anomalies[0]["evidence"]["anomaly_count"] = len(drift_flags)
anomalies[0]["evidence"]["results_df"] = drift_flags[
    ['date', 'pincode', 'district', 'state', 'age_ratio', 'ratio_zscore']
]

# ------------------------------------------------
# ANOMALY 2 (REVISED): Impossible Population Growth
# ------------------------------------------------
df_temp = df_temp.sort_values(['district', 'date'])
df_temp['prev_count'] = (
    df_temp.groupby('district')['demo_age_5_17'].shift(1)
)

df_temp['growth_rate'] = (
    (df_temp['demo_age_5_17'] - df_temp['prev_count']) /
    (df_temp['prev_count'] + epsilon)
)

# Data-driven extreme growth threshold
growth_threshold = df_temp['growth_rate'].quantile(0.9)

growth_flags = df_temp[
    (df_temp['prev_count'] >= 50) &              # avoid tiny-base explosions
    (df_temp['growth_rate'] > growth_threshold)
].copy()

anomalies[1]["evidence"]["anomaly_count"] = len(growth_flags)
anomalies[1]["evidence"]["results_df"] = growth_flags[
    ['date', 'district', 'state', 'demo_age_5_17', 'growth_rate']
]

# ------------------------------------------------
# ANOMALY 3 (REVISED): Spatial-Temporal Stagnation
# ------------------------------------------------
df_temp = df_temp.sort_values(['pincode', 'date'])

# Rolling range instead of std (robust for integer counts)
df_temp['rolling_range'] = (
    df_temp
    .groupby('pincode')['demo_age_5_17']
    .transform(lambda x: x.rolling(window=5).max() - x.rolling(window=5).min())
)

stagnant_flags = df_temp[
    df_temp['rolling_range'] < 2     # identical or Â±1 variation over 5 days
].copy()

anomalies[2]["evidence"]["anomaly_count"] = len(stagnant_flags)
anomalies[2]["evidence"]["results_df"] = stagnant_flags[
    ['date', 'pincode', 'district', 'state', 'demo_age_5_17']
]


# ------------------------------------------------
# ANOMALY 4: Geographical Multi-Mapping
# ------------------------------------------------
geo_counts = (
    df_demographic
    .groupby('pincode')['district']
    .nunique()
    .reset_index()
)
geo_counts.columns = ['pincode', 'district_count']

multi_mapped_pincodes = geo_counts[
    geo_counts['district_count'] > 2
]['pincode']

geo_flags = df_demographic[
    df_demographic['pincode'].isin(multi_mapped_pincodes)
].copy()

anomalies[3]["evidence"]["anomaly_count"] = len(geo_flags)
anomalies[3]["evidence"]["results_df"] = geo_flags[
    ['pincode', 'district', 'state']
]

# ------------------------------------------------
# ANOMALY 5 (FINAL): Cross-District Mass Imbalance
# ------------------------------------------------

district_daily = (
    df_demographic
    .groupby(['state', 'district', 'date'], as_index=False)
    .agg({
        'demo_age_5_17': 'sum',
        'demo_age_17_': 'sum'
    })
)

district_daily['district_total'] = (
    district_daily['demo_age_5_17'] +
    district_daily['demo_age_17_']
)

district_daily = district_daily.sort_values(['state', 'district', 'date'])
district_daily['prev_total'] = (
    district_daily.groupby(['state', 'district'])['district_total'].shift(1)
)

district_daily['delta'] = (
    district_daily['district_total'] -
    district_daily['prev_total']
)

state_daily = (
    district_daily
    .groupby(['state', 'date'], as_index=False)
    .agg(
        state_delta=('delta', 'sum'),
        state_abs_delta=('delta', lambda x: np.abs(x).sum())
    )
)

district_daily = district_daily.merge(
    state_daily,
    on=['state', 'date'],
    how='left'
)

district_daily['conserved'] = (
    (district_daily['state_abs_delta'] > 0) &
    (np.abs(district_daily['state_delta']) /
     (district_daily['state_abs_delta'] + epsilon) < 0.10)
)

district_daily['contribution_ratio'] = (
    np.abs(district_daily['delta']) /
    (district_daily['state_abs_delta'] + epsilon)
)

# Require BOTH gainers and losers on the same state-day
leakage_flags = []

for (state, date), grp in district_daily.groupby(['state', 'date']):
    if not grp['conserved'].iloc[0]:
        continue

    if (grp['delta'] > 0).any() and (grp['delta'] < 0).any():
        flagged = grp[grp['contribution_ratio'] > 0.30]
        leakage_flags.append(flagged)

leakage_df = pd.concat(leakage_flags) if leakage_flags else pd.DataFrame()

anomalies[4]["evidence"]["anomaly_count"] = len(leakage_df)
anomalies[4]["evidence"]["results_df"] = leakage_df[
    ['date', 'state', 'district', 'district_total', 'delta', 'contribution_ratio']
]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# --- 1. AGGREGATE ALL 4 TYPES WITH GEOGRAPHY ---
all_flags_list = []
for a in anomalies:
    res = a['evidence']['results_df']
    if res is not None and not res.empty:
        df_f = res.copy()
        df_f['anomaly_type'] = a['name']
        cols_to_keep = ['date', 'state', 'district', 'pincode', 'anomaly_type']
        available_cols = [c for c in cols_to_keep if c in df_f.columns]
        all_flags_list.append(df_f[available_cols])

if all_flags_list:
    df_all_flags = pd.concat(all_flags_list)

    # Identify the Top 4 States by total anomaly count
    top_4_states = df_all_flags['state'].value_counts().head(4).index.tolist()
    print(f"Generating Critical Audits for: {', '.join(top_4_states)}")

    for target_state in top_4_states:
        # --- 2. IDENTIFY DOUBLE-FAILURES FOR CURRENT STATE ---
        state_df = df_all_flags[df_all_flags['state'] == target_state]

        # Districts failing more than one distinct type of check
        failure_diversity = state_df.groupby('district')['anomaly_type'].nunique()
        critical_districts = failure_diversity[failure_diversity > 1].index.tolist()
        df_critical = state_df[state_df['district'].isin(critical_districts)]

        # --- 3. FINAL SUMMARY DASHBOARD PER STATE ---
        sns.set_theme(style="whitegrid")
        fig, axes = plt.subplots(1, 2, figsize=(18, 7))
        fig.suptitle(f"Critical Data Audit: {target_state}", fontsize=18, fontweight='bold')

        # Plot 1: Stacked Bar of Critical Districts (Multi-Type Failures)
        if not df_critical.empty:
            top_crit = df_critical.groupby(['district', 'anomaly_type']).size().unstack(fill_value=0)
            top_crit['sum'] = top_crit.sum(axis=1)
            top_crit = top_crit.sort_values('sum', ascending=False).head(15).drop(columns='sum')

            top_crit.plot(kind='bar', stacked=True, ax=axes[0], colormap='rocket')
            axes[0].set_title(f"Top 15 Districts with Multiple Anomaly Types", fontsize=14)
            axes[0].set_ylabel("Flag Count")
        else:
            axes[0].text(0.5, 0.5, "No multi-type failures detected", ha='center')

        # Plot 2: Anomaly Composition for the State
        state_comp = state_df['anomaly_type'].value_counts()
        sns.barplot(x=state_comp.values, y=state_comp.index, hue=state_comp.index, ax=axes[1], palette="viridis", legend=False)
        axes[1].set_title(f"Overall Anomaly Breakdown: {target_state}", fontsize=14)

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

        # --- 4. EXPORT CRITICAL EVIDENCE ---
        filename = f"critical_audit_{target_state.replace(' ', '_')}.csv"
        df_critical.to_csv(filename, index=False)
        print(f"Audit for {target_state} Complete. {len(df_critical)} critical rows exported to {filename}\n" + "-"*50)
else:
    print("No anomalies detected in any category.")
```
